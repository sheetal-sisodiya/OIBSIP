```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
df=datasets.load_iris()
```

```python
df.data
```

```python
df.target
```

```python
df.target_names
```

```python
df.data.shape
```

```python
x=df.data
```

```python
x.shape
```

```python
y=df.target
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_std=sc.fit_transform(x)
x_std
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.80)
```

```python
pca_x=pca.fit_transform(x_std)
```

```python
pca_x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(pca_x,y,train_size=0.20)
```

```python
x_train.shape,x_test.shape
```

###Logistic Regression & Random Forestare best fitted model for this dataset classification

1. logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_1=LogisticRegression()
```

```python
model_1.fit(x_train,y_train)
```

```python
y_pred_1=model_1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
model_answer=accuracy_score(y_pred_1,y_test)*100
```

```python
model_answer
```

2.Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_2=RandomForestClassifier(n_estimators=50,max_depth=3)
```

```python
model_2.fit(x_train,y_train)
```

```python
y_pred_2=model_2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred_2,y_test)*100
```

```python

param_grid={
    'n_estimators':[50,100,200],
    'max_depth':[1,2,3],
    'min_samples_split':[2,5,7],
    'min_samples_leaf':[1,2,4],
    'criterion':['gini','entropy']
}
3*3*3*3*2
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
base_model=RandomForestClassifier()
```

```python
grid_search=GridSearchCV(estimator=base_model,param_grid=param_grid)
```

```python
grid_search.fit(x_train,y_train)
```

```python
print(grid_search.best_params_)
```

```python
final_model=RandomForestClassifier(n_estimators=50,max_depth=2,min_samples_leaf=1,min_samples_split=7,criterion='gini')
```

```python
final_model.fit(x_train,y_train)
```

```python
y_pred_final=final_model.predict(x_test)
```

```python
accuracy_score(y_pred_final,y_test)*100
```
